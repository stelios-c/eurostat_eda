arrival and departure. But what is CAF, PAS_BRD, PAS_CRD, ST_PAS? See [metadata](https://ec.europa.eu/eurostat/cache/metadata/en/avia_pa_esms.htm)
> In the tables of the sub-domain "Transport measurement - Passengers", data are broken down by passengers on board (arrivals, departures and total), passengers carried (arrivals, departures and total) and passenger commercial air flights (arrival, departures and total). Additionally, the tables of collection "Detailed air transport by reporting country and routes" provide data on seats available (arrival, departures and total). The data is presented at monthly, quarterly and annual level.

```python
(data
 .query("freq=='A' and unit=='FLIGHT' and tra_meas=='CAF_PAS'")
 .dropna(axis='columns',how='all')
 .rename(columns={r'airp_pr\TIME_PERIOD':'airp_pr'})
 .assign(country_dep=lambda df: df.airp_pr.str.split('_',expand=True)[0],
         airport_dep=lambda df: df.airp_pr.str.split('_',expand=True)[1],
         country_arr= lambda df: df.airp_pr.str.split('_',expand=True)[2],
         airport_arr= lambda df: df.airp_pr.str.split('_',expand=True)[3]
)
 .query('country_dep=="FI" and country_arr=="FI"')
 .dropna(axis=0,how='all')
 #.T
 #.plot()

)
```
